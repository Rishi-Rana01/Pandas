```python
import pandas as pd
```

### Remove Rows
One way to deal with empty cells is to remove rows that contain empty cells.
This is usually OK, since data sets can be very big, and removing a few rows will not have a big impact on the result.

```python
df = pd.read_csv('data.csv')
new_df = df.dropna()

print("New data frame with no empty cell : \n",new_df.to_string())
```

### dropna() method returns a new DataFrame, and will not change the original.


```python
df = pd.read_csv('data.csv')
df.dropna(inplace = True) # change the original DataFrame
print("Data frame after removing empty cells : \n",df.to_string())
```

#### Replace Empty Values

```python
df = pd.read_csv('data.csv')
df.fillna(9999, inplace=True) # replace empty values with 130
print("Data frame after replacing empty cells : \n",df.to_string())
```

### Replace Only For Specified Columns

```python
df= pd.read_csv('data.csv')
df.fillna({'Calories': 8888}, inplace=True) # replace empty values in 'Calories' column with 8888
print("Data frame after replacing empty cells in 'Calories' column : \n",df.to_string())
```

### Replace Using Mean, Median, or Mode
Pandas uses the mean() median() and mode() methods to calculate the respective values for a specified column:


```python
df = pd.read_csv('data.csv')
x = df["Calories"].mean()  # mean value of 'Calories' column
print(x)

df["Calories"] = df["Calories"].fillna(x)  # replace empty values with the mean
print("Data frame after replacing empty cells in 'Calories' column with mean :\n" , df)
```

```python

df = pd.read_csv('data.csv')
x = df["Calories"].median()
print(x)

df.fillna({"Calories": x}, inplace=True)
print("Data frame after replacing empty cells in 'Calories' column with median :\n" , df)
```

### Mode

```python
df = pd.read_csv('data.csv')
x = df["Calories"].mode()[0]
print(x)

df.fillna({"Calories": x}, inplace=True)
print("Data frame after replacing empty cells in 'Calories' column with mode :\n" , df)
```
